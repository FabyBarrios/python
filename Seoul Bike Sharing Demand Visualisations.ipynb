**Abstract:**  The dataset contains count of public bikes rented at each hour in Seoul Bike haring System with the corresponding Weather data and Holidays information



**Attribute Information:**

**Date:** year-month-day

**Rented Bike count:** Count of bikes rented at each hour

**Hour:** Hour of he day

**Temperature:** Temperature in Celsius

**Humidity:** %

**Windspeed:** m/s

**Visibility:** 10m

**Dew point temperature:** Celsius

**Solar radiation:** MJ/m2

**Rainfall:** mm

**Snowfall:** cm

**Seasons:** Winter, Spring, Summer, Autumn

**Holiday:** Holiday/No holiday

**Functional Day:** NoFunc(Non Functional Hours), Fun(Functional hours)

```python
#Converting the Date column in Datetime Dtype
df.Date = pd.to_datetime(df.Date)
df
```

```python
#Breaking Down the Date into 3 Components
df['Day']=df['Date'].dt.day

df["Month"]=df.Date.dt.month_name().str[:3]

df['Year']=df['Date'].dt.year
```

```python
df['Day of the week'] = df.Date.dt.dayofweek
df['Day of the week'] = df['Day of the week'].map({0: "Monday",1: "Tuesday",2: "Wednesday" ,3: "Thursday",4:"Friday",5:"Saturday",6:"Sunday"})
df['Day of the week'] = pd.Categorical(df['Day of the week'], categories=
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],
    ordered=True)
df
```
